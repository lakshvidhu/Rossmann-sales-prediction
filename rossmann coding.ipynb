```python
# Concatenate the datasets
combined_inputs = pd.concat([train_inputs, val_inputs, test_inputs], axis=0)

# Apply pd.get_dummies to the combined dataset
combined_dummies = pd.get_dummies(combined_inputs[['DayOfWeek', 'Promo', 'StateHoliday', 'StoreType', 'Assortment']])

# Ensure encoded_cols corresponds to the columns in combined_dummies
encoded_cols = list(combined_dummies.columns)

# Split the datasets back
train_inputs[encoded_cols] = combined_dummies[:len(train_inputs)]
val_inputs[encoded_cols] = combined_dummies[len(train_inputs): len(train_inputs) + len(val_inputs)]
test_inputs[encoded_cols] = combined_dummies[len(train_inputs) + len(val_inputs):]

# Drop the original categorical columns if needed
train_inputs = train_inputs.drop(['DayOfWeek', 'Promo', 'StateHoliday', 'StoreType', 'Assortment'], axis=1)
val_inputs = val_inputs.drop(['DayOfWeek', 'Promo', 'StateHoliday', 'StoreType', 'Assortment'], axis=1)
test_inputs = test_inputs.drop(['DayOfWeek', 'Promo', 'StateHoliday', 'StoreType', 'Assortment'], axis=1)
```

```python
# Import necessary libraries
from sklearn.preprocessing import OneHotEncoder

# Assuming 'categorical_cols' is a list of columns to be one-hot encoded
# Ensure the column names are correct and present in the DataFrame
print(train_inputs.columns)

# Check if all columns in categorical_cols are present in train_inputs
missing_cols = set(categorical_cols) - set(train_inputs.columns)
if missing_cols:
    print(f"Columns {missing_cols} are missing in train_inputs. Please check the column names.")

# Convert categorical columns to strings if needed
train_inputs[categorical_cols] = train_inputs[categorical_cols].astype(str)

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore').fit(train_inputs[categorical_cols])

# Get feature names from the encoder
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))

# Transform and replace the original categorical columns
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])

# Perform similar steps for val_inputs and test_inputs
val_inputs[categorical_cols] = val_inputs[categorical_cols].astype(str)
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])

test_inputs[categorical_cols] = test_inputs[categorical_cols].astype(str)
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
X_train = train_inputs[numeric_cols + encoded_cols]
X_val = val_inputs[numeric_cols + encoded_cols]
X_test = test_inputs[numeric_cols + encoded_cols]
```

```python
def return_mean(inputs):
    return np.full(len(inputs), merged_train_data.Sales.mean())
```

```python
train_preds = return_mean(X_train)
train_preds
```

```python
from sklearn.metrics import mean_squared_error
print("Train RMSE:",(mean_squared_error(train_preds, train_targets, squared=False)))
print("Validation RMSE:", (mean_squared_error(return_mean(X_val), val_targets, squared=False)))
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, SGDRegressor
```

```python
def try_model(model):
    model.fit(X_train, train_targets)
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    train_rmse = mean_squared_error(train_targets, train_preds, squared=False)
    val_rmse = mean_squared_error(val_targets, val_preds, squared=False)
    return train_rmse, val_rmse
```

```python
print("LinearRegression RMSE:", (try_model(LinearRegression())))
print("Ridge RMSE:", (try_model(Ridge())))
print("Lasso RMSE:", (try_model(Lasso())))
print("ElasticNet RMSE:", (try_model(ElasticNet())))
print("SGDRegressor RMSE:", (try_model(SGDRegressor())))


```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree

tree = DecisionTreeRegressor(random_state=42)
try_model(tree)
```
